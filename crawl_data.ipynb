```python
def write(url):

    def find(x,string):
        obj=re.compile(x,re.S)
        result=obj.finditer(string)
        for it in result:
            # print(it.group('name'))
            return it.group('name')

    import requests
    import re
    import pandas as pd
    # url='https://www.hockey-reference.com/players/d/dewarto01.html'

    headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}

    resp=requests.get(url,headers=headers)
    # print(resp.text)

    name=find(r'"name": "(?P<name>.*?)"',resp.text)
    picture_url=find(r'"contentUrl": "(?P<name>.*?)"',resp.text)
    position=find('<p><strong>Position</strong>: (?P<name>.*?)&nbsp;&bull;&nbsp;    <strong>Shoots<',resp.text)
    shoots=find(r'<strong>Shoots</strong>: (?P<name>.*?)</p>',resp.text)
    try:
        shoots=shoots.strip()
    except:pass
    height=find(r'"height": { "@type": "QuantitativeValue", "value": "(?P<name>.*?)" }',resp.text)
    try:
        height=height.replace('-',"'")
    except:pass
    weight=find(r'"weight": { "@type": "QuantitativeValue", "value": "(?P<name>.*?)" }',resp.text)
    birthPLace=find(r'"birthPlace": "(?P<name>.*?)"',resp.text)
    try:
        birthPLace=birthPLace.replace('&nbsp;',' ')
    except:pass
    birthdate=find(r'"birthDate": "(?P<name>.*?)"',resp.text)
    died=find(r'data-death="(?P<name>.*?)">',resp.text)
    halloffame=find(r'Hall of Fame</strong>: Inducted as Player&nbsp;(?P<name>.*?) ',resp.text)
    try:
        halloffame='Inducted as Player'+halloffame
    except:pass
    dit={0:name,1:picture_url,2:position,3:shoots,4:height,5:weight,6:birthPLace,7:birthdate,8:died,9:halloffame}
    print(dit)

    df = pd.DataFrame(dit, index=[0])
    df.to_csv('player_data.csv', mode='a', header=0, index=False, encoding='UTF8')
import pandas as pd
import time


df=pd.read_csv('href.csv')
hrefs=(df['href'].to_list())

dit={0:'name',1:'picture_url',2:'position',3:'shoots',4:'height',5:'weight',6:'birthPLace',7:'birthdate',8:'died',9:'hall of fame'}
df = pd.DataFrame(dit, index=[0])
df.to_csv('player_data.csv', mode='a', header=0, index=False, encoding='UTF8')

for href in hrefs:
    print('HREF IS HERE:',href)
    write(href)
    time.sleep(0.1)
    # break
```
